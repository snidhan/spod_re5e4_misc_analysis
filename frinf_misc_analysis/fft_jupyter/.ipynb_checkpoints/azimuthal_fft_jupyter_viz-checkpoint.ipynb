```python
####......................Reading the grid file............................

nx, index, x, xe, xc = io.readgrid('/home/sheel/Work/projects/spod_re5e4/grid/frinf/x3_grid.in')
nr, index, r, re, rc = io.readgrid('/home/sheel/Work/projects/spod_re5e4/grid/frinf/x1_grid.in')
ntheta = 258
nr_trunc = nr - 10 + 1

print("Number of grid points in x direction", nx)
print("Number of grid points in r direction", nr)
print("Number of grid points in theta direction", ntheta)
```

```python
###..................Read series of slice cuts available......................

nstart = 2300000
nend   = 2613000
stride = 1000


path = '/home/sheel/Work2/projects_data/spod_re5e4/frinf/data_files_uniform/x_D_40/up/'

file_header = 'up_'


nit  =  int((nend-nstart)/stride + 1)
print('nit = ', nit)

## Looping over the window for plane files available

data_full = np.zeros((nr_trunc, ntheta, nit))

data_m = np.zeros((nr_trunc,ntheta))
    

for it in range(0,nit):

    file_number  =  nstart + it*stride
    filename   =  path  + file_header + '%08d'%file_number + '_40_uniform_pchip.res'
    data = np.fromfile(filename, dtype='float64')
    data = np.reshape(data, (ntheta, nr_trunc))
    data_full[:,:,it] = data.T
    data_m = data_m + (1.0/nit)*data.T
```

```python
###..............Data range...................................................

print(np.amax(data_m))
print(np.amin(data_m))
print(data_m.shape)
data_full = data_full - np.repeat(data_m[:, :, np.newaxis], nit, axis=2)
print(np.amax(data_full))
print(np.amin(data_full))
print(data_full.shape)
```

```python
###.............Centering the data value...................................

# data_centered = 0.5*(data_full[:,1:-1,:] + data_full[:,0:-2,:])               ## Works for the azimuthal velocity field

data_centered_x = 0.5*(data_full[1:-1,1:-1,:] + data_full[0:-2,1:-1,:])         ## Works for the radial velocity field
data_centered = np.zeros((nr-10+1, ntheta-2, nit))
data_centered[0,:,:] = data_centered_x[0,:,:]
data_centered[-1,:,:] = data_centered_x[-1,:,:]

print(np.amax(data_centered))
print(np.amin(data_centered))
print(np.shape(data_centered))
```

```python
###.............. Performing azimuthal fourier transform of the velocity field........................

data_centered_fft = 1/(ntheta-2)*np.fft.fftn(data_centered, axes = (1,))

###............... Extracting the modal content from FFTed content....................................
data_centered_fft_m2 = data_centered_fft[:,2,:]
data_centered_fft_m1 = data_centered_fft[:,1,:]
data_centered_fft_m0 = data_centered_fft[:,0,:]
```

```python
###.............. Multiplying by e^{im\theta} in the azimuthal direction......................

theta = np.linspace(0+np.pi/256, 2*np.pi-np.pi/(ntheta-2), ntheta-2)

azimuthal_variation_m2 = np.exp(1j*2*theta)
azimuthal_variation_m1 = np.exp(1j*1*theta)
azimuthal_variation_m0 = np.exp(1j*0*theta)
```

```python
### Creating 2D data slice from the FFTed components........................

data_2d_fft = np.zeros((356,ntheta-2,np.shape(data_centered_fft_m2)[1]), dtype=complex)

for k in range(0, np.shape(data_centered_fft_m2)[1]):
    data_2d_fft[:,:,k] = np.reshape(data_centered_fft_m2[:,k],(356,1))*np.reshape(azimuthal_variation_m2, (1, 256))

    
data_2d_fft_full = np.zeros((356,ntheta,np.shape(data_centered_fft_m2)[1]), dtype=complex)
data_2d_fft_full[:,1:-1,:] = data_2d_fft
data_2d_fft_full[:,0,:] = data_2d_fft[:,-1,:]
data_2d_fft_full[:,-1,:] = data_2d_fft[:,1,:]
```

```python
###..................Creating the meshgrid...................................................

v = 5
idx = (np.abs(v - rc)).argmin()
print(idx)

azimuths = np.linspace(-(0.5)*2*np.pi/(ntheta-2),2*np.pi+0.5*2*np.pi/(ntheta-2) , ntheta) 
R, THETA = np.meshgrid(rc[0:idx],azimuths)

R = R.T
THETA = THETA.T

print(R.shape)
print(THETA.shape)
```

```python
###.......................Visualizing averaged velocity field.............................

#fig = plt.figure(dpi=300)
fig, ax1 = plt.subplots(subplot_kw=dict(projection='polar'), dpi = 200)
cs1 = ax1.contourf(THETA, R, np.real(data_2d_fft_full[0:idx,:,5]), cmap = 'viridis', levels=np.linspace(-0.005,0.005,20), extend = 'both')
cbaxes = fig.add_axes([0.8, 0.8, 0.1, 0.04])
cb = plt.colorbar(cs1,cax = cbaxes,orientation = 'horizontal', ticks = [-0.005,0.005], label = "Re($\hat{u}_{r}$)")
ax1.set_aspect(1.0)
plt.show()
```

```python
##..............................Animating the velocity field.............................

fps = 10 # frame per sec
frn = nit # frame number of the animation

fig, ax = plt.subplots(subplot_kw=dict(projection='polar'), dpi = 200)
cs1     = ax.contourf(THETA, R, np.real(data_2d_fft_full[0:idx,:,0]), cmap = 'viridis', levels=np.linspace(-0.005, 0.005,20), extend = 'both')
cbaxes  = fig.add_axes([0.1, 0.1, 0.2, 0.04])
cb = plt.colorbar(cs1,cax = cbaxes,orientation = 'horizontal', ticks = [-0.005,0.005], label = "Re($\hat{u}_{\\theta}$)")
plt.show()

def update_plot(frn):
    ax.clear()
    cs1     = ax.contourf(THETA, R, np.real(data_2d_fft_full[0:idx,:,frn]), cmap = 'viridis', levels=np.linspace(-0.005,0.005,20), extend = 'both')
    cb = plt.colorbar(cs1,cax = cbaxes,orientation = 'horizontal', ticks = [-0.005,0.005], label = "Re($\hat{u}_{\\theta}$)")
    ax.set_aspect(1.0)

ani = animation.FuncAnimation(fig, update_plot, frn, interval=200)

fn = 'v_animation_m2_frinf_x_D_40'
ani.save(fn+'.mp4',writer='ffmpeg',fps=fps)
```
